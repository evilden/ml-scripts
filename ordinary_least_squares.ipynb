```python
import numpy as np
from matplotlib import pyplot as plt

X = 2*np.random.rand(100)
y = 4 + 3*X + np.random.randn(100)

plt.style.use(['ggplot'])
plt.figure(figsize=(20, 8))
plt.plot(X, y, 'b.')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,2,0,15])
```

OLS

```python
X1 = np.c_[np.ones((100)),X]
W = np.linalg.inv(X1.T @ X1) @ X1.T @ y

X_est = np.array([0, 2])
X_est_1 = np.c_[np.ones((2)), X_est]
y_est = X_est_1 @ W

plt.figure(figsize=(20, 8))
plt.plot(X_est, y_est, 'r-')
plt.plot(X, y, 'b.')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,2,0,15])
```
